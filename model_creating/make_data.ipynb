```python
import os

dirs = os.listdir('D:/PythonWork/OldZmiy/HomePlantsAssistant/files')
all_files = []
for d in dirs:
    files = os.listdir(f'D:/PythonWork/OldZmiy/HomePlantsAssistant/files/{d}')
    for f in files:
        all_files.append(f'D:/PythonWork/OldZmiy/HomePlantsAssistant/files/{d}/{f}')
all_files
```

```python
import json

# Выходной JSON файл
output_file = "data.json"

# Словарь для хранения данных
data = []
new_data = []

for filename in all_files:
    # Читаем текст из файла
    print(filename)
    try:
        with open(filename, 'r', encoding='utf-8') as file:
            content = file.read()
            
        # Формируем запись для каждого документа
        record = {
            "id": filename.split('.')[0],  # Берём имя файла без расширения как ID
            "title": filename.replace(".txt", "utf-8"),  # Удаляем расширение .txt
            "text": content
        }
        
        new_record = {
            "id": filename.split('.')[0],
            "content": content
        }
        
        data.append(record)
        new_data.append(new_record)
        
    except Exception:
        with open(filename, 'r', encoding='windows-1251') as file:
            content = file.read()
            
        # Формируем запись для каждого документа
        record = {
            "id": filename.split('.')[0],  # Берём имя файла без расширения как ID
            "title": filename.replace(".txt", ""),  # Удаляем расширение .txt
            "text": content
        }
        data.append(record)
        
        new_record = {
            "id": filename.split('.')[0],
            "content": content
        }
        
        data.append(record)
        new_data.append(new_record)

# Запись данных в JSON файл
with open(output_file, 'w', encoding='utf-8') as outfile:
    json.dump(data, outfile, ensure_ascii=False, indent=4)
    
with open('new_data.json', 'w', encoding='utf-8') as outfile2:
    json.dump(new_data, outfile2, ensure_ascii=False, indent=4)

print(f"Данные записаны в {output_file}")
```
